# <p><span style="color:black">Extracting only the necessary columns</span></p>

```r
datavar2 <- data.frame(G = datavar$G,A = datavar$A)
head(datavar2)
```

# <p><span style="color:black">Adding data for Patrick as per the specification. </span></p>

```r
datavar3 <- rbind(datavar2, c(285, 462))
```

# <p><span style="color:black">Creating linear models for: <br> 1. Original data. <br> 2. Passing through wayne. <br> 3. Passing through Patrick</span></p>

```r
linearModelVar_orig <- lm(A ~ G, datavar2)
linearModelVar_wayne <- lm(I(A-1963) ~ I(G-894) + 0, data=datavar2)
linearModelVar_pat<- lm(I(A-462) ~ I(G-285) + 0, data=datavar3)
# summary(linearModelVar)
```

```r
y_int_wayne <- 1963 - (coef(linearModelVar_wayne)*894)
y_int_pat <- 462 - (coef(linearModelVar_pat)*285) # g - 285 a - 462
```

# <p><span style="color:black">Plots for Original data and line passing through Wayne. </span></p>

```r
#wayne 
plot(A~G,datavar2)
abline(linearModelVar_orig)
abline(y_int_wayne,coef(linearModelVar_wayne), data=datavar2)
```

# <p><span style="color:black">Plots for original data and line passing through Patrick.</span></p>

```r
#wayne 
plot(A~G,datavar3)
abline(linearModelVar_orig)
abline(y_int_pat,coef(linearModelVar_pat), data=datavar3)
```

# <p><span style="color:black">Making the Wayne plot to go through origin i.e.the model passes through both the point and origin.</span></p>

```r
#multiplication factor calculation 
old_slope <- coef(linearModelVar_wayne)
needed_slope <- (1963-0)/(894-0)
multiplication_factor <- needed_slope/old_slope
multiplication_factor

linearModelVar_wayne2 <- lm(I((A-1963)*multiplication_factor) ~ I(G-894) + 0 , data=datavar2)
plot(I(A)~I(G),datavar2)
abline(0,coef(linearModelVar_wayne2), data=datavar2)
```

# <p><span style="color:black">Making the Patrick plot to go through origin i.e.the model passes through both the point and origin.</span></p>

```r
#multiplication factor calculation 
old_slope <- coef(linearModelVar_pat)
needed_slope <- (462-0)/(285-0)
multiplication_factor <- needed_slope/old_slope
multiplication_factor

linearModelVar_pat2 <- lm((I(A-462)*multiplication_factor) ~ I(G-285) + 0 , data=datavar3)
plot(I(A)~I(G),datavar3,xlim=c(0,max(G)+50),ylim=c(0,max(A)+50))
abline(0,coef(linearModelVar_pat2), data=datavar2)
```

# <p><span style="color:black">Making the original plot to go through origin i.e.the model passes through origin.</span></p>

```r
linearModelVar_orig_2 <- lm(A ~ G + 0, datavar2)
plot(A~G,datavar2,xlim=c(0,max(G)+50),ylim=c(0,max(A)+50))
abline(linearModelVar_orig_2)
```

# <p><span style="color:black">Implications : <br><br>High R squared -  means the model explains a lot of variation within the data<br> AND <br> Low P - model is significant</span></p>

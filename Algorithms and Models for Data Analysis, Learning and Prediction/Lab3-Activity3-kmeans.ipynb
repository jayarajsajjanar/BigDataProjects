# <p><span style="color:black">As per the instructions in the lab description, 5 variables are considered for clustering. Only the variables that are of similar nature are considered. The 5 variables have values 1,2,8 and 9. 8 and 9 are filtered out.   </span></p>

```r
df4 <- df[, c("eminuse","intmob","home4nw","smart1","cable1")]
# head(df4)
df5 <- na.omit(df4)
```

```r
df5
```

# <p><span style="color:black">Filtering out any values that are not 1 or 2. </span></p>

```r
#eminuse	intmob	home4nw	smart1	cable1
df6 <- df5[df5$eminuse < 8 & df5$intmob < 8 & df5$home4nw < 8 & df5$smart1 < 8 & df5$cable1 < 8  ,]
# table(df5)
```

```r
table(df6$eminuse)
table(df6$intmob)
table(df6$home4nw)
table(df6$smart1)
table(df6$cable1 )
```

```r
#reference - http://www.learnbymarketing.com/tutorials/k-means-clustering-in-r-example/
k_set <- 2:20

iter <-100 
avg_tot_ss <-integer(length(k_set)) 

for(k in k_set){ 
     v_tot_ss <-integer(iter) 
     #finds average of tot.withinss of 100 tries
     for(i in 1:iter){
         temp <-kmeans(df6,centers=k) 
         v_tot_ss[i] <-temp$tot.withinss
        }
     avg_tot_ss[k-1] <-mean(v_tot_ss) 
}

plot(k_set,avg_tot_ss,type="b", main="avg_tot_ss vs k", ylab="avg_tot_ss", xlab="K")
```

# <p><span style="color:black">Just like the example in the vignette, there is no clear elbow. And since dramatic drops in avg_tot_ss happen till about 5 clusters (also in comparison with k=3). We can choose k=5 clusters as optimal. </span></p>

```r
final_result <-kmeans(df6,centers=5) 
```

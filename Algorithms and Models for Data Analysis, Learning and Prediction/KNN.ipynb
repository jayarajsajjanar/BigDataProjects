# <p><span style="color:black"> Only the columns that contain numerical data are considered for the training process. <br> <br>The columns that contained numerical data are normalised to ensure that only a few variables do not dominate in the knn finding process.<br><br> Normalisation in done in such a way that all the columns contained values from 0-1</span></p>

```r
max_job <- max(data$Job, na.rm = TRUE)
min_job <- min(data$Job, na.rm = TRUE)
normalized_job = (data$Job-min_job)/(max_job-min_job)

max_Age <- max(data$Age, na.rm = TRUE)
min_Age <- min(data$Age, na.rm = TRUE)
normalized_age = (data$Age-min_Age)/(max_Age-min_Age)

max_cred_amt <- max(data$Credit.amount, na.rm = TRUE)
min_cred_amt <- min(data$Credit.amount, na.rm = TRUE)
normalized_cred_amt = (data$Credit.amount-min_cred_amt)/(max_cred_amt-min_cred_amt)

max_duration <- max(data$Duration, na.rm = TRUE)
min_duration <- min(data$Duration, na.rm = TRUE)
normalized_duration = (data$Duration-min_duration)/(max_duration-min_duration)
```

```r
# normalized_data <- cbind(normalized_age,normalized_job)
normalized_data <- cbind(normalized_age,normalized_job,normalized_cred_amt,normalized_duration)
```

```r
normalized_data
```

# <p><span style="color:black"> Miscalculation error rate is calculated for different values of k and with different ratios of training and test data.</span></p>

```r
results = data.frame()
```

```r
for (k in 1:20) {
    
#     print(k)
    col_num=2
    results[k,1] = k
    for (j in 6:8){
        
        #The percentage of data to be taken as train/test is taken from for loop.
        len_train <- ((10*j)/100)*nrow(normalized_data)
        tot <- nrow(normalized_data)
        len_test <- tot - len_train
       
        #Splitting the data into train and test as per the percentage.
        normalized_data_train <- normalized_data[0:len_train,]
        normalized_data_test <- normalized_data[(len_train+1):tot,]
        cl_train <- data[0:len_train,]$Credit.Risks
        cl_test <- data[(len_train+1):tot,]$Credit.Risks
        
        #Applying the knn function.
        predicted.labels <- knn(normalized_data_train, normalized_data_test, cl_train, k)

        #Calculating the misclassification rate.
        num.incorrect.labels <- sum(predicted.labels != cl_test)
        misclassification.rate <- num.incorrect.labels / len_test
        misclassification.rate
        
        #Forming the results dataframe.
        results[k,col_num]=misclassification.rate
        col_num = col_num+1
    }
    
}
colnames(results) <- c("k","60% Train Set","70% Train Set","80% Train Set")
results
```

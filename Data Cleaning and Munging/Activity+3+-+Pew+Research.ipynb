# <p><span style="color:Red"> Q2.  Where do people who replied Excellent to the communities they live in work?</span></p>

```r
df_q1_1 <- filter(df, q1 == 1) 
emptyp_cnt_exc_locality <- df_q1_1 %>% group_by(emptype1) %>% summarise(emptype_count = n())
emptype_count <- as.vector(emptyp_cnt_exc_locality$emptype_count)
```

# <p><span style="color:red">Q3. Who are more inclined to use internet/email? male or female?</span></p>

```r
# Who uses internet/email more? male or female?
# df_q1_1 <- filter(df, q1 == 1) 
count_per_sex_vec <- df %>% group_by(sex,eminuse) %>% summarise(count_per_sex = n())
count_per_sex_df <- as.data.frame(count_per_sex_vec)
count_per_sex_df_sex_1 <- slice(count_per_sex_df,1:2)
count_per_sex_df_sex_2 <- slice(count_per_sex_df,3:4)
```

```r
count_per_sex_vec_slice_sex_1 <- as.vector(count_per_sex_df_sex_1$count_per_sex)
count_per_sex_vec_slice_sex_2 <- as.vector(count_per_sex_df_sex_2$count_per_sex)
```

```r
par(mfrow=c(2,2))

slices1 <- count_per_sex_vec_slice_sex_1
lbls1 <- c("Yes","No")
pct1 <- round(slices1/sum(slices1)*100)

lbls1 <- paste(lbls1, pct1) # add percents to labels
lbls1 <- paste(lbls1,"%",sep="") # ad % to labels
#---------------------------------------------------------------------------
slices2 <- count_per_sex_vec_slice_sex_2
lbls2 <- c("Yes","No")
pct2 <- round(slices2/sum(slices2)*100)

lbls2 <- paste(lbls2, pct2) # add percents to labels
lbls2 <- paste(lbls2,"%",sep="") # ad % to labels


print ("Who are more inclined to use internet/email? male or female?")
pie(slices1, labels = lbls1, col=rainbow(length(lbls1)), main="Male")
pie(slices2, labels = lbls2, col=rainbow(length(lbls2)), main="Female")
```

# <p><span style="color:blue">  Since more `females` per 100 use internet/email, they are more inclined. </span></p>

# <p><span style="color:red">Q4. Which ethic group has more percent of people working in a Non Profit Organisation?</span></p>

```r
df_q1_1 <- filter(df, emptype1 == 1) 
npo_count_race <- df_q1_1 %>% group_by(race) %>% summarise(npo_count = n())
race_count <- df %>% group_by(race) %>% summarise(race_count = n())
calc_percent <- inner_join(npo_count_race, race_count, by = "race")
calc_percent2 <- mutate(calc_percent, percent_wise = (npo_count/race_count)*100)
```

```r
percent_final_npo_race <- as.vector(calc_percent2$percent_wise)
```

```r
barplot(percent_final_npo_race, names.arg = c("White","Black","AsnPcf","MxdRc","Native","Other","Dntknow","Rfusd"),
       ylab="Percentage" , xlab="Race/Ethic Group")
```

# <p><span style="color:blue">  We can infer that `Asians-Pacific Islanders` have more percent of people working at a non profit organisation </span></p>

# <p><span style="color:red">Q6. Which ethic group has least divorce rates?</span></p>

```r
df_divorced <- filter(df, marital == 3) 
# df_divorced
divorce_count_race <- df_divorced %>% group_by(race) %>% summarise(divorce_count = n())
race_count <- df %>% group_by(race) %>% summarise(race_count = n())
calc_percent <- left_join(race_count, divorce_count_race, by = "race")
calc_percent2 <- mutate(calc_percent, percent_wise = (divorce_count/race_count)*100)
calc_percent2[is.na(calc_percent2)] <- 0
```

```r
percent_final_race_divorced <- as.vector(calc_percent2$percent_wise)
```

```r
data=data.frame(group= c("White","Black","AsnPcf","MxdRc","Native","Other","Dntknow","Rfusd"), value=percent_final_race_divorced)
```

```r
library(ggplot2)
```

```r
ggplot(data, aes(x = group, y = value ,fill = group)) + 
    geom_bar(width = 0.85, stat="identity") +    
    coord_polar(theta = "y") +    
    xlab("") + ylab("") +
    ylim(c(0,75)) + 
    geom_text(data = data, hjust = 1, size = 3, aes(x = group, y = 0, label = group)) +
    theme(legend.position = "none" , axis.text.y = element_blank() , axis.ticks = element_blank())
```

# <p><span style="color:blue">  We can infer that `Native Americans` have least divorce rates. </span></p>

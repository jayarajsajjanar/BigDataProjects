```r
wo_na_locs <- long_lat_google[complete.cases(long_lat_google),]
```

## Final map will display how cities reacted on a topic quantitatively. In order to achieve it, longs and lats are grouped and their <em> count </em>s are extracted.  This method works because <em> geocode </em> accepts locations/cities names and returns their lat/long. <br>

## Grouping by lon-lat <br>

```r
# grouping the long/lat and extracting their counts/frequencies.
long_lat_non_dup <- count(wo_na_locs, c("lon", "lat"))
```

```r
length(long_lat_non_dup$lon)
```

```r
#storing in a temporary variable. 
x <- long_lat_non_dup #x has lon lat freq with length
```

```r
#The long and lats are spread all over the world.. This code ignores them and considers only the USA ones.
boxed_long_lat <- x[((x$lon) < (-75))& ((x$lon) > (-130)) & (x$lat) > 22 & (x$lat) < 50, ]
```

```r
normalised_freq <- (boxed_long_lat$freq - min(boxed_long_lat$freq))*20/(max(boxed_long_lat$freq)-min(boxed_long_lat$freq))
```

```r
ggplot_formatted <- data.frame(
  lon = boxed_long_lat$lon,
  lat = boxed_long_lat$lat,
#   freq = boxed_long_lat$freq,
    freq = normalised_freq,
  stringsAsFactors = FALSE
  ) 
```

```r
title_message = paste("USA cities reacting to - ", topic, " - Tweets are grouped by cities. ")
```

```r
states <- map_data("state")
```

### Bigger yellow circles imply more number of tweets from that city.  i.e relative sizes not absolute<br>

```r
ggplot(data = states) + labs(title = title_message) + 
  geom_polygon( aes(x = long, y = lat,group = group), color = "white") + 
  coord_fixed(1.3) +
  guides(fill=FALSE) + 
#   geom_point(data = temp3, aes(x = temp3$Var1, y = temp3$Var2), color = "yellow", size = 4) 
#   geom_point(data = ggplot_formatted, aes(x = ggplot_formatted$lon, y = ggplot_formatted$lat), color = "yellow", size = ggplot_formatted$freq/10)
geom_point(data = ggplot_formatted, aes(x = ggplot_formatted$lon, y = ggplot_formatted$lat), color = "yellow", size = ggplot_formatted$freq)
```
